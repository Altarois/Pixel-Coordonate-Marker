```python
import pygame
import numpy as np
import cv2

#ouverture de l'image
img = cv2.imread('resultat.jpg')
img_out = img.copy()

#initialisation de Pygame
pygame.init()

#paramétres de la fenetre
white = (255,255,255)
black= (0,0,0)
value=0

size = [1000, 700]
screen=pygame.display.set_mode(size)

pygame.display.set_caption("TP Isat")

#on charge l'image en mémoire
image = pygame.image.load(r'C:\Users\à modifier\resultat.png')

done=False 

clock = pygame.time.Clock()

#boucle qui execute les events dans la fenetre
while done==False:
      for event in pygame.event.get():
            if event.type == pygame.QUIT :
                  done=True  
      #on enregistre les positions de la souris dans 2 variables
      pos=pygame.mouse.get_pos()
      x=pos[0]  
      y=pos[1]
        
      screen.fill(white)
    
      x=pos[0]  
      y=pos[1]  
    
      #ouverture de l'image et dessin des lignes et de la position des pixels
      screen.blit(image, (0, 0))
      pygame.draw.line(screen , black , [x-10,0],[x-10,y-10],2)
      pygame.draw.line(screen , black , [x-10,800],[x-10,y-10],2)
      pygame.draw.line(screen , black , [0,y-10],[x-10,y-10],2)
      pygame.draw.line(screen , black , [800,y-10],[x-10,y-10],2)
      font=pygame.font.Font(None , 25)
      
      text=font.render('{}, {}'.format(x, y),True,white)
      
      screen.blit(text,[x+10,y])
      
      #evenement de clic pour le découpage de l'image
      if event.type == pygame.MOUSEBUTTONUP:
          if event.button == 1:
                    #incrementation +2 et sauvegarde de la position de la souris au premier clic
                    value+=2
                    mx,my=pygame.mouse.get_pos()
                    print(pygame.mouse.get_pos())
                    
           #evenement du 2 eme clic          
          if event.button == 3:
             #incrémentation de +1 et sauvegarde de la position de la souris au 2 eme clic  
             value+=1
             nx,ny=pygame.mouse.get_pos()
             print('2 eme click',pygame.mouse.get_pos())
             
      #si value devient impair alors sa veut dire qu on a effectuer 2 clics 
      if(value%2 == 1):
                #affichage de l'image selon les clics effectuer
                im=img_out[mx:nx,my:ny]      
                cv2.imshow('image',im)
                cv2.waitKey(0) 
                cv2.destroyAllWindows()
                value+=1
      
        
      pygame.display.flip()
      #sert a definir le taux de raffraichissement par seconde de la fenetre 
      clock.tick(60)  

pygame.quit()
```
